```python
import requests
import pandas as pd
```

```python
def get_coord(x):
    r3 = requests.get(x).json()
    coords = r3['features'][0]['geometry']
    print(coords)
    return coords
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/jbuffer/flood-actions/main/data/flood-data.csv')
```

```python
get_coord(df.polygon_url[0])
```
